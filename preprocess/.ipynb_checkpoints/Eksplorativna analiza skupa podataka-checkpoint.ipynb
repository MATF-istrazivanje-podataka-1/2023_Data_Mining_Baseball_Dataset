```python
from scipy.io import arff
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

plt.style.use('ggplot')
```

```python
data = arff.loadarff(r'..\dataset\dataset_189_baseball.arff')
df = pd.DataFrame(data[0])
df.info()
```

```python
def decode(df, columns):
    for column in columns:
        df[column] = df[column].map(lambda x : x.decode())
```

```python
columns = ['Player', 'Hall_of_Fame', 'Position']
decode(df, columns)
df.head(20)
```

```python
df = df.drop('Player', axis=1)
```

```python
df[['Position']]
```

```python
df = pd.get_dummies(df, prefix='', prefix_sep='',columns=['Position'], dtype=float)
```
