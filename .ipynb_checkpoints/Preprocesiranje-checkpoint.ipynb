```python
def histogram(df, nazivkolone, tehnika):
    df[nazivkolone].plot(kind='hist', color='red', bins=40, alpha=0.5, edgecolor='black')
    plt.xlabel(nazivkolone + ' ' + tehnika, fontsize=14)
    plt.ylabel('Frequency', fontsize=14)
    
    mean = df[nazivkolone].mean()
    p5 = df[nazivkolone].quantile(0.05)
    p95 = df[nazivkolone].quantile(0.95)
    

    plt.axvline(mean, color='blue', label='mean')
    plt.axvline(p5, color='green', label='5th Percentile')
    plt.axvline(p95, color='purple', label='95th Percentile')

    plt.legend()
    plt.show()
```

```python
histogram(df, 'Strikeouts', 'default')
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.ensemble import RandomForestRegressor
```

```python
df.columns
```

```python
imptr = IterativeImputer(RandomForestRegressor(), max_iter=100, tol=0.01, random_state=0)

df1 = dfnum[['Number_seasons', 'Games_played', 'At_bats', 'Runs', 'Hits', 'Doubles',
       'Triples', 'Home_runs', 'RBIs', 'Walks', 'Strikeouts',
       'Batting_average', 'On_base_pct', 'Slugging_pct', 'Fielding_ave',
       'Position']]

dfpredi = pd.DataFrame(imptr.fit_transform(df1), columns = df1.columns)
dfpredi.head()
```

```python
histogram(df, 'Strikeouts', 'default')
histogram(dfpredi, 'Strikeouts', 'imputation')
```

```python
dfnum['Strikeouts'] = dfpredi['Strikeouts'].copy()
dfnum.isna().any()
```

```python
def z_score(X):
    return (X - X.mean()) / X.std()
```

```python
from sklearn.preprocessing import MinMaxScaler

columns_to_normalize = ['Games_played', 'At_bats', 'Runs', 'Hits', 'Doubles',
       'Triples', 'Home_runs', 'RBIs', 'Walks', 'Strikeouts']

for kolona in columns_to_normalize:
    dfnum[kolona] = z_score(dfnum[kolona]).copy()

dfnum.head(20)
```
