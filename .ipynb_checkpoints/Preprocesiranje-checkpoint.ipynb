## Izbacivanje nepoznatih instanci

```python
def histogram(df, nazivkolone, tehnika, xlimmin, xlimmax):
    df[nazivkolone].plot(kind='hist', bins=30, color='red', alpha=0.5, edgecolor='black')
    plt.xlabel(nazivkolone + ' ' + tehnika, fontsize=14)
    plt.ylabel('Frequency', fontsize=14)
    plt.xlim(xlimmin, xlimmax)
    
    mean = df[nazivkolone].mean()
    p5 = df[nazivkolone].quantile(0.05)
    p95 = df[nazivkolone].quantile(0.95)
    

    plt.axvline(mean, color='blue', label='mean')
    plt.axvline(p5, color='green', label='5th Percentile')
    plt.axvline(p95, color='purple', label='95th Percentile')

    plt.legend()
    plt.show()
```

```python
histogram(df.dropna(), 'Strikeouts', 'izbacivanje null vrednosti', 0, 2700)
```

```python
from sklearn.ensemble import IsolationForest
```

```python
IsolationForest().fit(df[['Number_seasons']]).predict(df[['Number_seasons']])
```

```python
from sklearn.covariance import EllipticEnvelope
EllipticEnvelope().fit(df[['Number_seasons']]).predict(df[['Number_seasons']])
```

```python
def outliers_find(df,x):
    q1  =  df[x].quantile(.25)
    q3  =  df[x].quantile(.75)
    iqr = q3-q1
    df['Traditional']  = np.where(df[[x]]<(q1-1.5*iqr),-1,
                         np.where(df[[x]]>(q3+1.5*iqr),-1,1))
    df['Isolation Forest'] = IsolationForest().fit(df[[x]]).predict(df[[x]])
    df['Eliptic Envelope'] =EllipticEnvelope().fit(df[['Number_seasons']]).predict(df[['Number_seasons']])     
    return df
```

```python
outliers_find(df, x='Number_seasons')
```
